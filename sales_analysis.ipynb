```python
df = pd.read_csv(
    "C:/Users/Avanti Y/Desktop/CODING/projects/sales_analysis/sales_data_sample.csv",
    header=None,
    encoding="latin1"
)

df.columns = [
    "ORDERNUMBER", "QUANTITYORDERED", "PRICEEACH", "ORDERLINENUMBER", "SALES",
    "ORDERDATE", "STATUS", "QTR_ID", "MONTH_ID", "YEAR_ID", "PRODUCTLINE",
    "MSRP", "PRODUCTCODE", "CUSTOMERNAME", "PHONE", "ADDRESSLINE1",
    "ADDRESSLINE2", "CITY", "STATE", "POSTALCODE", "COUNTRY", "TERRITORY",
    "CONTACTLASTNAME", "CONTACTFIRSTNAME", "DEALSIZE"
]
```

```python
df.head()
df.info()
```

```python
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"], errors="coerce")
df["SALES"] = pd.to_numeric(df["SALES"], errors="coerce")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["ORDERDATE"] = pd.to_datetime(df["ORDERDATE"], errors="coerce")
df["SALES"] = pd.to_numeric(df["SALES"], errors="coerce")
df["QUANTITYORDERED"] = pd.to_numeric(df["QUANTITYORDERED"], errors="coerce")
df = df.drop_duplicates()
```

```python
df["YEAR"] = df["ORDERDATE"].dt.year
df["MONTH"] = df["ORDERDATE"].dt.month
df["YEAR_MONTH"] = df["ORDERDATE"].dt.to_period("M").astype(str)
```

```python
total_sales = df["SALES"].sum()
total_orders = df["ORDERNUMBER"].nunique()
avg_order_value = df.groupby("ORDERNUMBER")["SALES"].sum().mean()

monthly_sales = df.groupby("YEAR_MONTH")["SALES"].sum().reset_index()
top_products = df.groupby("PRODUCTLINE")["SALES"].sum().sort_values(ascending=False).reset_index()
top_customers = df.groupby("CUSTOMERNAME")["SALES"].sum().sort_values(ascending=False).head(10).reset_index()
sales_by_country = df.groupby("COUNTRY")["SALES"].sum().sort_values(ascending=False).reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=monthly_sales, x="YEAR_MONTH", y="SALES", marker="o")
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(data=top_products.head(10), x="SALES", y="PRODUCTLINE")
plt.title("Top Product Lines")
plt.tight_layout()
plt.show()
```

Monthly sales trend

```python
monthly_sales = (
    df.groupby("YEAR_MONTH")["SALES"]
      .sum()
      .reset_index()
      .sort_values("YEAR_MONTH")
)

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly_sales, x="YEAR_MONTH", y="SALES", marker="o")
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.ylabel("Total Sales")
plt.xlabel("Year-Month")
plt.tight_layout()
plt.show()
```

Top product lines

```python
top_productlines = (
    df.groupby("PRODUCTLINE")["SALES"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

plt.figure(figsize=(8, 5))
sns.barplot(data=top_productlines, x="SALES", y="PRODUCTLINE")
plt.title("Sales by Product Line")
plt.xlabel("Total Sales")
plt.tight_layout()
plt.show()
```

Top customers

```python
top_customers = (
    df.groupby("CUSTOMERNAME")["SALES"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
      .reset_index()
)

plt.figure(figsize=(8, 5))
sns.barplot(data=top_customers, x="SALES", y="CUSTOMERNAME")
plt.title("Top 10 Customers by Sales")
plt.xlabel("Total Sales")
plt.tight_layout()
plt.show()
```

 Sales by country

```python
sales_by_country = (
    df.groupby("COUNTRY")["SALES"]
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

plt.figure(figsize=(8, 5))
sns.barplot(data=sales_by_country, x="SALES", y="COUNTRY")
plt.title("Sales by Country")
plt.xlabel("Total Sales")
plt.tight_layout()
plt.show()
```
